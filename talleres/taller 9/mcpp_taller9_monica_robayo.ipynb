28.Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

```python
def percent(word,text):
    fdist=FreqDist(text)
    w_word=fdist[word]     
    total=len(text)     
    percent=w_word/total     
    print( str(word), "appears", str(percent) + "% in ", str(text))
```

```python
percent("JOIN",text5)
```

```python
percent("lots",text5)
```

4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
# importamos libreria y se codifican loops para captar la frecuencia de los objetivos (targets)
import nltk 
freq = nltk.ConditionalFreqDist((target, fileid[:4]) 
  for fileid in nltk.corpus.state_union.fileids() 
  for v in nltk.corpus.state_union.words(fileid) 
                                
#luego se especifica que elemntos se tomarán de los discursos, esto es: (target) women, men, people                           
  for target in ['men', 'women', 'people'] 
                                
# por ultimo se tomaca cada instancia del objeto(target) cuya forma comienza en una minúscula
  if v.lower().startswith(target)) 

freq.plot() ;
```

Respuesta: Se evidencia que antes de 1970 la palabra mujer casi no se usaba, pero desde dicho año se ha usado más o menos como el término hombre. El termino people ha sido más nombrado para los últimos años. 

11. Investigate the table of modal distributions and look for other patterns. Try to explain them in terms of your own impressionistic understanding of the different genres. Can you find other closed classes of words that exhibit significant differences across different genres?

```python
from nltk.corpus import brown
brown.categories()
```

```python
freq2 = nltk.ConditionalFreqDist((genre, word)
                               
       for genre in brown.categories()
       for word in brown.words(categories=genre))

genres = ['editorial',
 'fiction',
 'government',
 'hobbies',
 'humor',
 'learned',
 'lore',
 'mystery',
 'news',
 'religion',
 'reviews',
 'romance',
 'science_fiction' ]

general_words = ['can', 'could', 'may', 'might', 'must', 'will']

freq2.tabulate(conditions=genres, samples=general_words)
```

Se evidencia que se usan más los verbos modales en las categorias Hobbies, Government y Learned. Además se evidencia que en las categorías Humor y Science_fiction  se usan muy poco. 

```python
freq3 = nltk.ConditionalFreqDist((genre1, word)
                                 
       for genre1 in brown.categories()
       for word in brown.words(categories=genre1))

genres1 = ['editorial',
 'fiction',
 'government',
 'hobbies',
 'humor',
 'learned',
 'lore',
 'mystery',
 'news',
 'religion',
 'reviews',
 'romance',
 'science_fiction' ]

general_words1 = ['what', 'where', 'when', 'why', 'which']

freq3.tabulate(conditions=genres1, samples=general_words1)
```

Se evidencia que en la categoría Learned la palabra "which" es la más usada. Además la palabra que menos es usada en las categorías es Why y la más usada en which.
